**Q4.** This question looks at a time series of the number of active oil drilling rigs in the United States over time. The data comes from the Energy Information Agency.

1. Load `./data/drilling_rigs.csv` and examine the data. How many observations? How many variables? Are numeric variables correctly read in by Pandas, or will some variables have to be typecast/coerced? Explain clearly how these data need to be cleaned.

```python
df = pd.read_csv('https://github.com/alaplace17/EDA/blob/main/assignment/data/drilling_rigs.csv?raw=true') # Local file

df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

There are 623 observations, and 10 variables. 
Some variables did not read in correctly, so they need to be coerced and cleaned. 

```python
df['time'] = pd.to_datetime(df['Month'], format='mixed', errors='coerce')
df['time'].head()
df.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Replace "Not Available" with NaN and coerce columns to numeric
    df.replace("Not Available", pd.NA, inplace=True)
    columns_to_convert = [
        "Crude Oil Rotary Rigs in Operation, Total (Number of Rigs)",
        "Natural Gas Rotary Rigs in Operation, Total (Number of Rigs)",
        "Crude Oil and Natural Gas Rotary Rigs in Operation, Horizontal Trajectory (Number of Rigs)",
        "Crude Oil and Natural Gas Rotary Rigs in Operation, Directional Trajectory (Number of Rigs)",
        "Crude Oil and Natural Gas Rotary Rigs in Operation, Vertical Trajectory (Number of Rigs)",
        "Active Well Service Rig Count (Number of Rigs)"
    ]
    df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

3. Let's look at `Active Well Service Rig Count (Number of Rigs)`, which is the total number of rigs over time. Make a line plot of this time series. Describe what you see.

4. Instead of levels, we want to look at change over time. Compute the first difference of  `Active Well Service Rig Count (Number of Rigs)` and plot it over time. Describe what you see.

```python
# Compute the first difference of 'Active Well Service Rig Count (Number of Rigs)'
df_clean['Rig Count Change'] = df_clean['Active Well Service Rig Count (Number of Rigs)'].diff()

# Plot the first difference over time
plt.figure(figsize=(10, 6))
plt.plot(df_clean['time'], df_clean['Rig Count Change'])
plt.xlabel('Time')
plt.ylabel('Change in Active Well Service Rig Count (Number of Rigs)')
plt.title('Change in Total Number of Rigs Over Time')
plt.grid(True)
plt.show()
```

There were some large spikes and large drops espeically around 1986 ad 2001. In 2020, another drop, and since then the change has been pretty steadily close to zero.
